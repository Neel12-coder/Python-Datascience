# Matplotlib

Matplotlib is multi-platform data visualization tool built on the Numpy and Sidepy frameworks

```python
!pip install matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
y
```

```python
plt.plot(x,y,"y-") #third arg is to change color
#plt.show() to print o/p
```

```python
plt.plot(x,y,"r-")
plt.xlabel("XLABEL")
plt.ylabel("YLABEL")
plt.title("TITLE")
```

```python
#multiple plots inside same canvas
plt.subplot(1,2,1)  #nrows,ncol,no.of plt
plt.plot(x,y,"r")

plt.subplot(1,2,2)
plt.plot(y,x,'y')
```

```python
#OBJECT ORIENTED APPROACH

fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left,right, width, height
axes.plot(x,y)
axes.set_xlabel("X LABEL")
axes.set_ylabel("Y LABEL")
axes.set_title("OBJECT ORIENTED GRAPH")
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)
fig.tight_layout()#for no overlap
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title("first")
axes[1].plot(y,x)
axes[1].set_title("second")
```

```python
fig = plt.figure(figsize=(8,2)) #defines width and ht for plot .. next arg is dpi (dots per inches)
ax = fig.add_axes([0,0,1,1])
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig = plt.figure(figsize=(8,2)) #defines width and ht for plot .. next arg is dpi (dots per inches)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
#Saving a Figure

fig.savefig("myfig.jpg",dpi=200)
```

```python
#Adding Legend in Grapg

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label="abc")
ax.plot(y,x,label="pqr")
ax.legend(loc=2) #loc is location
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label="abc",color="yellow",linewidth=2,alpha = 1.5,linestyle="--",marker="o",markersize="10",
        markerfacecolor="blue",markeredgewidth="10",markeredgecolor="red")#use HEX CODE for custom color
#for transparency of plot use alpha
#for linewidth u can use lw
# line style can be -- or :, '.' is not a valid value for ls; supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
# marker for points highligh
```

#Control Over Axis Appearance / Plot Range

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color="red",lw="2",ls="--")
ax.set_xlim([0,2])
ax.set_ylim([0,4])
```
