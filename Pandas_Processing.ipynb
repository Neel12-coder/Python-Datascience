```python
import pandas as pd
```

```python
listings = pd.read_csv("listings.csv")
```

```python
listings.head()
```

```python
listings.shape
```

```python
listings[["price","host_response_rate","host_acceptance_rate"]].head()
```

```python
listings["price"]=listings["price"].str[1:].str.replace(",","").astype(float) 
```

```python
listings["price"]
```

```python
listings.describe
```

```python
listings.dtypes
```

```python
listings["host_response_rate"]=listings["host_response_rate"].str[:-1].astype(float)/100
```

```python
listings["host_response_rate"].head(10)
```

```python
listings["host_acceptance_rate"]=listings["host_acceptance_rate"].str[:-1].astype(float)/100
```

```python
listings["host_acceptance_rate"].head(10)
```

```python
listings[["bathrooms_text"]].head()
```

```python
#bathroom_text contains number of bathrooms and bathroom type
listings["bathroom_numbers"] = listings["bathrooms_text"].str.split(" ",expand=True)[0]
```

```python
listings["bathroom_type"] = listings["bathrooms_text"].str.split(" ",expand=True)[1]
```
